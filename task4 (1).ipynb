```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    """
    This function validates the recharge amount.
    Conditions:
    - Amount must be >= 50
    - Amount must be one of the valid plans
    """
    
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans:", valid_plans)
        return False
    
    return True


# Retry system using while loop
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: ₹"))
        
        if validate_recharge(recharge_amount):
            print("Recharge successful! ✅")
            break
        else:
            print("Please try again.\n")
    
    except ValueError:
        print("Invalid input. Please enter numbers only.\n")
```

```python
# Prbolem 2: Inventory Reorder Alert System

def check_inventory(products):
    """
    This function checks stock levels.
    If stock is less than 15 → Reorder Alert
    Otherwise → Stock OK
    """
    
    for product, stock in products.items():
        if stock < 15:
            print(product, "→ Reorder Alert (Stock:", stock, ")")
        else:
            print(product, "→ Stock OK (Stock:", stock, ")")


inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Salt": 25
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    """
    This function:
    - Calculates the average marks
    - Returns 'Pass' if average >= 50
    - Returns 'Fail' otherwise
    """
    
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Sample input
student_marks = [60, 55, 40, 70, 50]

# Function call
avg, result = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", result)



```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak):
    """
    Calculates cab fare based on:
    - Base fare = ₹50
    - ₹12 per km
    - 25% extra during peak hours
    """
    
    base_fare = 50
    per_km_rate = 12
    
    total_fare = base_fare + (distance * per_km_rate)
    
    # Add 25% extra if peak hour
    if peak.lower() == "yes":
        total_fare += total_fare * 0.25
    
    return total_fare


# Retry system
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")
        
        fare = calculate_fare(distance, peak)
        
        print("Total Fare: ₹", fare)
        
        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            print("Thank you for using Cab Fare Estimator")
            break
    
    except ValueError:
        print("Invalid input. Please enter numeric value for distance.\n")
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    """
    This function:
    - Calculates attendance percentage
    - Returns 'Eligible' if percentage >= 75
    - Returns 'Not Eligible' otherwise
    """
    
    present_count = 0
    
    # Count number of 'P'
    for day in attendance_list:
        if day == "P":
            present_count += 1
    
    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100
    
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Sample Input
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Function Call
percent, result = check_attendance(attendance)

print("Attendance Percentage:", percent)
print("Status:", result)
```
