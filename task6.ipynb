```python
#  program1: Smart Parking Lot Management System

def manage_parking(capacity, logs):
    
    current_vehicles = 0
    peak_vehicles = 0
    
    # process vehicle logs
    for entry in logs:
        
        if entry == "IN":
            current_vehicles += 1
        elif entry == "OUT" and current_vehicles > 0:
            current_vehicles -= 1
        
        # track peak usage
        if current_vehicles > peak_vehicles:
            peak_vehicles = current_vehicles
    
    print("Currently Parked Vehicles:", current_vehicles)
    
    if current_vehicles > capacity:
        print("Parking Status: Full - Capacity Exceeded")
    else:
        print("Parking Status: Available")


# sample input
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(parking_capacity, vehicle_logs)
```

```python
#  program 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    
    # base time (5 minutes per km)
    time = distance * 5
    
    # traffic delay
    if traffic.lower() == "high":
        time += 10
    elif traffic.lower() == "medium":
        time += 5
    
    # weather delay
    if weather.lower() == "rainy":
        time += 5
    elif weather.lower() == "stormy":
        time += 15
    
    print("Estimated Delivery Time:", time, "minutes")


# sample input
distance_km = 8
traffic_level = "High"
weather_condition = "Rainy"

estimate_delivery_time(distance_km, traffic_level, weather_condition)
```

```python
# program 3:  Movie Theatre Seat Occupancy Analyzer

def analyze_occupancy(total_seats, booked_seats):
    
    # count booked seats
    booked_count = len(booked_seats)
    
    # calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", occupancy, "%")
    
    # check show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# sample input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

analyze_occupancy(total_seats, booked_seats)
```

```python
# program 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    
    # calculate total using loop
    total = 0
    for value in cpu_readings:
        total += value
    
    # calculate average
    average = total / len(cpu_readings)
    
    print("Average CPU Load:", average, "%")
    
    # classify load
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# sample input
cpu_data = [45, 60, 70, 85, 90]

classify_server_load(cpu_data)
```

```python
# program 5:  Smart Classroom Resource Usage Monitor

def monitor_resources(resource_usage):
    
    overused = []
    
    # checking each resource
    for resource, hours in resource_usage.items():
        
        # assume overuse if usage is more than 8 hours
        if hours > 8:
            overused.append(resource)
    
    # display results
    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# sample input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage_data)
```

```python
#  program 6: Online Event Registration Capacity Controller

def manage_registrations(capacity, registrations):
    
    # check if registrations exceed capacity
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# sample input
event_capacity = 100
total_registrations = 105

manage_registrations(event_capacity, total_registrations)
```
