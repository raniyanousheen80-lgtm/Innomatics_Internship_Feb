```python
# Problem statement 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find employees who have the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Display result
print("Top Performers Eligible for Bonus:", 
      ", ".join(top_performers), 
      "(Score:", highest_score, ")")

```

```python
# Problem  2: Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation (basic handling)
for ch in ",.!?":
    query = query.replace(ch, "")

words = query.split()

word_count = {}

# Count frequency
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Display words searched more than once
result = {}

for word, count in word_count.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # Check if even (valid reading)
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):", valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_users = len(emails)

# Calculate and display percentage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage}%")
```

```python
# Problem 5: Sales Spike Detection

# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

for i in range(1, len(sales)):
    previous_day = sales[i - 1]
    current_day = sales[i]
    
    # Check if current day is 30% higher than previous day
    if current_day > previous_day * 1.30:
        print("Day", i + 1, ":", current_day)
```

```python
# Problem 6: Duplicate User ID Detection

# Input list
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store frequency of each user ID
frequency = {}

# Count occurrences
for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

# Display only duplicates
for user, count in frequency.items():
    if count > 1:
        print(user, "→", count, "times")
```
