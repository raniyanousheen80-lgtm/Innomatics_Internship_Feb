```python
# program 1:  Smart Light Controller

class SmartLight:
    
    # constructor
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # default status
    
    # turn ON
    def turn_on(self):
        self.status = "ON"
        print(self.name, "is", self.status)
    
    # method to turn OFF
    def turn_off(self):
        self.status = "OFF"
        print(self.name, "is", self.status)


# sample input
light = SmartLight("Bedroom Light")
light.turn_on()
```

```python
# program 2: Employee ID Card System

class Employee:
    
    # constructor
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
  
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# sample input
emp1 = Employee("Rahul", "EMP102", "IT")

emp1.display_id_card()
```

```python
#  program 3: Mobile Contact Record

class Contact:
    
    # constructor
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
   
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# sample input
contact1 = Contact("Anita", "9876543210")

contact1.display_contact()
```

```python
#  program 4: Product Price Tag Generator

class Product:
    
    # constructor
    def __init__(self, name, price):
        self.name = name
        self.price = price
    

    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# sample input
item = Product("Headphones", 2499)

item.print_price_tag()
```

```python
# program5:  Movie Rating Display System

class Movie:
    
    # constructor
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    def display_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# sample input
movie1 = Movie("Inception", 4.8)

movie1.display_details()
```

```python
# program 6 :  Delivery Address Manager

class Delivery:
    
    # constructor
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
  
    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# sample input
order1 = Delivery("Suman", "Hyderabad")

order1.display_delivery()
```
