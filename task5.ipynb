```python
#  program1: Social Media Post Engagement Analyzer

def check_post_engagement(likes):
    
    total = 0
    
    # adding all likes
    for i in likes:
        total = total + i
    
    print("Total Likes:", total)
    
    if total >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# example data
post_likes = [250, 300, 200, 300]

check_post_engagement(post_likes)
```

```python
#  program2: Medicine Stock Alert System

def check_medicine_stock(stock):
    
    print("Medicine Stock:", stock)
    
    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


# example value
medicine_stock = 6

check_medicine_stock(medicine_stock)
```

```python
# program3:Rainfall Adequacy Checker

def check_rainfall(rainfall_data, required_level):
    
    total = 0
    
    # calculate total rainfall
    for value in rainfall_data:
        total = total + value
    
    average = total / len(rainfall_data)
    
    print("Average Rainfall:", int(average))
    
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# example data
season_rainfall = [60, 75, 80, 70, 75]
required = 70

check_rainfall(season_rainfall, required)
```

```python
#  program4: Duplicate Account Detection

def check_duplicate_accounts(usernames):
    
    unique_users = set(usernames)
    
    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# example data
user_list = ["john123", "sara45", "john123", "alex99"]

check_duplicate_accounts(user_list)
```

```python
# program 6: Premium Crop Price Filter

def filter_premium_crops(prices):
    
    premium_list = []
    
    # check each price
    for price in prices:
        if price > 2000:
            premium_list.append(price)
    
    print("Premium Crops:", premium_list)


# example data
crop_prices = [1500, 2500, 1800, 3200, 2000]

filter_premium_crops(crop_prices)
```

```python
#  program 7 :Application Health Checker

def check_application_health(error_count):
    
    print("Error Count:", error_count)
    
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# example value
errors = 7

check_application_health(errors)
```

```python
#  program 8: Daily Transaction Limit Checker

def check_transaction(amount):
    
    daily_limit = 50000
    
    print("Transaction Amount:", amount)
    
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# example value
transaction_amount = 60000

check_transaction(transaction_amount)
```

```python
#  program 9: Student Attendance Eligibility System

def check_exam_eligibility(attendance_list):
    
    present_days = 0
    
    # count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    total_days = len(attendance_list)
    percentage = (present_days / total_days) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# example data
attendance = ["P", "P", "A", "P", "P"]

check_exam_eligibility(attendance)
```

```python
#  program 10: Smart Electricity Bill Analyzer

def calculate_bill(units):
    
    bill = 0
    
    # slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status


# example value
units_consumed = 250

total_bill, usage_status = calculate_bill(units_consumed)

print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
